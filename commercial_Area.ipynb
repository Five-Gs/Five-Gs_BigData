#### 부산광역시 상권 데이터만 추출

```python
busan = df[df['시도명'] == '부산광역시']
busan
```

```python
busan.shape
```

#### 사용할 칼럼만 추출

```python
# 사용할 column만 추출
columns = ['상권명', '시도명', '시군구명', '상권좌표수', '상권좌표']

busan = busan[columns].copy()
busan.head()
```

#### 한 행에 들어있는 여러가지 위경도 좌표 분리

```python
busan_loc = busan['상권좌표'].str.split('|')
busan_loc
```

```python
# busan_loc = busan_loc.reset_index(drop=True)
# busan_loc
```

```python
busan_loc.info()
```

```python
busan_loc[254]
```

```python
# 분리한 좌표를 데이터프레임의 칼럼으로 나타내기
busan_loc = pd.DataFrame(busan_loc.values.tolist(), index=busan.index)
busan_loc
```

```python
# 분리한 좌표를 데이터프레임의 행으로 나타내기
busan_df = pd.DataFrame(busan_loc.stack()).reset_index()
df_1 = busan_df.rename(columns={0:'좌표'})
df_1.pivot_table(index=['level_0','level_1','좌표'])
```

```python
df_1.index
```

```python
# level_0을 상권좌표의 인덱스로 설정
df_1.set_index('level_0', inplace=True)
```

```python
df_1
```

#### 분리한 좌표 데이터와 부산 상권 데이터와 결합

```python
busan_df = busan.join(df_1, how='inner')
busan_df
```

```python
# 필요없는 칼럼 제거
df_busan = busan_df.drop(['상권좌표', 'level_1'], axis=1)
```

```python
df_busan
```

```python
df_busan['좌표'].shape[0]
```

### 부산 상권 좌표를 지도에 마커로 표시

```python
m = folium.Map(location = [35.172883, 129.130644], zoom_start = 13)
nrow = 0

for i in range(df_busan['좌표'].shape[0]):
    try:
        for item in df_busan['좌표'][i].iteritems():
            subtime = item[1].split(',')
            # print(subtime[0], subtime[1])
            folium.Marker(location = [subtime[1], subtime[0]]).add_to(m)
    except:
        continue

m
```

### 상권좌표 마커를 개수로 지도에 표시

```python
m = folium.Map(location = [35.172883, 129.130644], zoom_start = 13)
cluster = MarkerCluster().add_to(m)
nrow = 0

for i in range(df_busan['좌표'].shape[0]):
    try:
        for item in df_busan['좌표'][i].iteritems():
            subtime = item[1].split(',')
            # print(subtime[0], subtime[1])
            folium.Marker(location = [subtime[1], subtime[0]]).add_to(cluster)
    except:
        continue

m
```

### 부산시 구별로 상권 밀집도 표현

```python
busan
```

#### 구별로 상권좌표수 개수 그룹화

```python
busan_group = busan.groupby(['시군구명']).sum()
busan_group
```

```python
# 지도에 나타내기 위해 '시군구명' 인덱스를 '시군구' 칼럼으로 추가
busan_group['시군구'] = busan_group.index 
```

```python
busan_group
```

#### 부산시 구별 간 경계선 데이터 불러오기

```python
geo_path = './data/busan_gu.json'

import json
geo_json = json.load(open(geo_path, encoding='utf-8'))
```

```python
geo_json['features'][0]['properties']['name']
```

```python
lat = 35.181784
long = 129.052852
(lat, long)
```

#### 상권 수를 기준으로 부산시 구 별 밀집도 시각화

```python
map = folium.Map([lat, long])

folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=busan_group,
    columns=['시군구', '상권좌표수'],
    key_on='feature.properties.name',
    fill_color='PuBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='상권 수'
).add_to(map)

map
```
